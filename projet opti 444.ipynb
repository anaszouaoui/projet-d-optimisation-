```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Jn
def Jn(u, A, b):
    return 0.5 * np.dot(u.T, np.dot(A , u)) - np.dot(b, u)     # calcule 1/2 uT*A*u - bT*u 

# la représentation pour n=2
n = 2
h = 1 / (n + 1)
A = (1 / h) * np.array([[2, -1], [-1, 2]])  
b = np.array([1, 1])  
# maillage de points
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
U, V = np.meshgrid(x, y)
Z = np.array([Jn(np.array([u, v]), A, b) for u, v in zip(U.flatten(), V.flatten())]).reshape(U.shape)

# Représentation 3D
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_surface(U, V, Z, cmap='viridis')
ax.set_xlabel('u1')
ax.set_ylabel('u2')
ax.set_zlabel('Jn(u)')
plt.show()
```

```python
# Vérification des valeurs propres
eigvals = np.linalg.eigvals(A)
print("Valeurs propres de A :", eigvals)
print("Définie positive :", np.all(eigvals > 0))
```
